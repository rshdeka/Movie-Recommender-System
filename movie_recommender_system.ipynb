### Feature Engineering

```python
# Let's explore the 'genres' column
movies['genres'].value_counts()
```

```python
import ast

def convert_genre(obj):
    genre = []
    for i in ast.literal_eval(obj):
        genre.append(i["name"])
    return genre

movies['genres'] = movies['genres'].apply(convert_genre)
```

```python
movies.head(3)
```

```python
# Let's explore the 'keywords' column
movies['keywords'].value_counts()
```

```python
def convert_keyword(obj):
    words = []
    for i in ast.literal_eval(obj):
        words.append(i["name"])
    return words

movies['keywords'] = movies['keywords'].apply(convert_keyword)
```

```python
movies.head(3)
```

```python
# Let's explore the 'cast' column
movies['cast'].value_counts()
```

```python
def convert_cast(obj):
    cast = []
    counter = 0
    for i in ast.literal_eval(obj):
        # We need the names of only top 3 actors
        if counter != 3:
            cast.append(i["name"])
            counter += 1
        else:
            break
    return cast

movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head(3)
```

```python
# Let's explore the 'crew' column
movies['crew'].value_counts()
```

```python
def convert_crew(obj):
    crew = []
    for i in ast.literal_eval(obj):
        # We need the name of the director only
        if i["job"] == "Director":
            crew.append(i["name"])
            break
    return crew

movies['director'] = movies['crew'].apply(convert_crew)
```

```python
movies = movies.drop(columns=['crew'], axis=1)
movies.head(3)
```

```python
# Convert the 'overview' column to list too
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(3)
```

```python
# The next step would be to strip all the spaces between names and keyword instances.
# This is done so that our vectorizer doesn't count the "Johnny" of "Johnny Depp" and "Johnny Galecki" as the same.
# Now when we remove space 'JohnnyDepp' and 'JohnnyGalecki' both are treated as single entities now. No confusion.
```

```python
# Remove spaces for the concerned columns
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['director'] = movies['director'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(3)
```

```python
# Concatenate these columns into a single column 'tags', which is a string that contains all the metadata that we want to feed to our vectorizer.
movies['tags'] = movies['overview']+movies['keywords']+movies['genres']+movies['cast']+movies['director']
movies.head(3)
```

```python
data = movies[['movie_id','title','tags']]
data.head(3)
```

```python
# Convert list to string
data['tags'] = data['tags'].apply(lambda x:" ".join(x))
data.head(3)
```

```python
# Convert tags to lowercase
data['tags'] = data['tags'].apply(lambda x:x.lower())
data.head(3)
```

```python
data['tags'][0]
```

### Using CountVectorizer() to extract features from text

```python
# CountVectorizer is used to convert a collection of text documents to a vector of term/token counts.
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features=5000,stop_words='english')

# Tokenize and build vocab
vectorizer.fit(data['tags'])

# Print the identified unique words along with their indices
print("Vocabulary: ", vectorizer.vocabulary_)
```

```python
# Encode document
vector = vectorizer.transform(data['tags'])

# Summarize encoded vector
print(vector.shape)
print(vector.toarray())
```

```python
vectorizer.get_feature_names()
```

### Using Stemming() to normalize words to their root forms

```python
# Clearly, there are some issues, like 'action' ~ 'actions' should be same
# So, we will use "Stemming" -> ["likes","liked","likely","liking"] -> ["like"]
```

```python
# Stemming is the process of producing morphological variants of a root/base word.
import nltk
from nltk.stem import PorterStemmer
 
ps = PorterStemmer()
 
def stem(text):
    list1 = []
    for i in text.split():
        # stem each word
        list1.append(ps.stem(i))
    
    # Convert list to string now
    return " ".join(list1)

data['tags'] = data['tags'].apply(stem)
```

### Calculate 'Cosine' Similarity of a movie with every other movie

```python
# Cosine Similarity is a measure of similarity, often used to measure document similarity in text analysis.
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vector)
print(similarity.shape)
```

#### Understand how to recommend movies

```python
# Movie Index -> Similarity matrix Index -> Sort -> Fetch movies
```

```python
# Get index of a movie
s1 = data[data['title'] == 'Batman Begins'].index[0]
s1
```

```python
# Get the list of cosine similarity scores for that particular movie with all movies.
dist = similarity[s1]
dist
```

```python
# When we will sort the movies, we will lose its index, but we need to maintain its index always
sorted(list(enumerate(dist)), reverse=True)
```

```python
# Sorting has been done acc. to index, which we don't want, we want ot sort acc. to similarity scores
sorted(list(enumerate(dist)), reverse=True, key=lambda x:x[1])
```

### Recommend Movies 

```python
def recommend(movie):
    # Get index of a movie
    movie_index = data[data['title'] == movie].index[0]
    # To get the distance with every other movie, we need to get its index in the similarity matrix
    distances = similarity[movie_index]
    # Sort the movies acc to similarity
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])
    # Fetch the top 10 similar movies
    movies_list = movies_list[1:11]
    # Get the movie indices and return the corresponding titles
    for i in movies_list:
        movie_indices = i[0]
        print(data['title'].iloc[movie_indices])
```

```python
recommend("The Avengers")
```
